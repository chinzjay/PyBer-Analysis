```python
%matplotlib inline

#Import dependancies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax=plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig=plt.figure()
ax=fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer fare by month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
#Create the plot
ax = plt.axes()
ax.plot(x_axis, y_axis, marker= "D", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
ax.set_xlabel('Date')
ax.set_ylabel("Fare($)")
# Create a title.
ax.set_title('PyBer Fare by Month')
# Set the y limit between 0 and 45.
ax.set_ylim(0,45)
# Add a grid.
ax.grid()
# Add the legend.
ax.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, color='red', label='Chicago')

plt.ylabel('Date')
plt.xlabel('Fare($)')
plt.gca().invert_yaxis()
plt.xlim(0,45)
plt.title('PyBer Fare by Month')
plt.legend()
```

```python

y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
fig, ax= plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
#Bubble chart
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color= 'skyblue', alpha=0.8, edgecolors='black',  linewidth=2, s=[i*5 for i in y_axis], label='Boston')
ax.set_ylabel('Date')
ax.set_xlabel('Fare($)')
ax.set_title('PyBer Fare by Month')
ax.set_xlim(0,50)
ax.invert_yaxis()
ax.legend()
```

```python
plt.pie(y_axis, labels=x_axis)
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, labels=x_axis, explode=explode_values, autopct='%.1f%%', colors=colors )
```
